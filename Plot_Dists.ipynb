```python
import numpy as np
import csv
from matplotlib import pyplot as plt
import scipy.stats as stats
from IPython.core.pylabtools import figsize
from nipy import load_image, save_image
from nipy.core.api import Image
import pymc3 as pm
import theano.tensor as T
%pylab inline
plt.style.use('ggplot')
```

```python
#Varying Both
label = ["Prior: 1 +/- 0.1", "Scanner 1: 0.9 +/- 0.1", "Scanner 2: 1.1 +/- 0.15"]
a = [100, 81, 53.77]
b = [0.01, 0.0111, 0.02045]
cols = ['r-', 'b-', 'g-']
fig, ax = plt.subplots(1,1)
for param in range(len(a)):
    X = np.linspace(stats.gamma.ppf(0.01, a[param], scale=b[param]),
                stats.gamma.ppf(0.99, a[param], scale=b[param]), 100) 
    ax.plot(X, stats.gamma.pdf(X, a[param], scale=b[param]),
         cols[param], lw=5, alpha=0.6, label=label[param])
fig.legend()
```

```python
#Varying Both
label = ["Prior: 0 +/- 0.01", "Scanner 1: 0.01 +/- 0.008", "Scanner 2: 0.02 +/- 0.004"]
a = [0, 0.01, 0.02]
b = [0.01, 0.008, 0.004]
cols = ['r-', 'b-', 'g-']
fig, ax = plt.subplots(1,1)
for param in range(len(a)):
    X = np.linspace(stats.norm.ppf(0.001, a[param], scale=b[param]),
                stats.norm.ppf(0.999, a[param], scale=b[param]), 220) 
    ax.plot(X, stats.norm.pdf(X, a[param], scale=b[param]),
         cols[param], lw=5, alpha=0.6, label=label[param])
fig.legend()
```
